```python
from langchain_community.embeddings import OllamaEmbeddings
```

## Ollama embeddings - Gemma:2b

```python
#by default it use llama2
embeddings = (
    OllamaEmbeddings(model = "gemma:2b") 
)
```

```python
embeddings
```

```python
text_vectors = embeddings.embed_documents(
    [
        "Data science isa multidisciplinary field that uses data to solve problems and inform decision-making." ,
        "It combines skills from mathematics, statistics, computer science, and other fields.",
    ]
)
```
